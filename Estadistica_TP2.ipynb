## Pregunta 1

1. Una variable aleatoria discreta $X$ puede tomar los valores 0, 1, 2 y 3. Las probabilidades para cada valor posible están dadas por la siguiente tabla:

$$
\begin{array}{|c|c|c|c|c|}
\hline
X & 0 & 1 & 2 & 3 \\
\hline
p & \frac{3\theta}{3} & \frac{6\theta}{3} & \frac{1-3\theta}{3} & \frac{2(1-3\theta)}{3} \\
\hline
\end{array}
$$

Si experimentalmente se obtienen los siguientes datos:
$(3, 3, 2, 1, 2, 3, 3, 0, 1, 1)$,
determine el valor de $\theta$ usando el método de máxima verosimilitud.

```python


# Parámetro a estimar
theta = sp.Symbol('theta', positive=True)

# Datos
datos = [3, 3, 2, 1, 2, 3, 3, 0, 1, 1]

# Conteo de frecuencias
n0 = datos.count(0)
n1 = datos.count(1)
n2 = datos.count(2)
n3 = datos.count(3)

# Probabilidades
p0 = (3 * theta) / 3
p1 = (6 * theta) / 3
p2 = (1 - 3 * theta) / 3
p3 = 2 * (1 - 3 * theta) / 3

# Multiplicamos la probabilidad de cada suceso elevada a la cantidad de veces que ocurrio
L = (p0**n0) * (p1**n1) * (p2**n2) * (p3**n3)
print(f"Función de Verosimilitud (theta):\n{sp.simplify(L)}\n")

# expand_log ayuda a separar los productos en sumas para facilitar la derivada
l_theta = sp.expand_log(sp.log(L), force=True)

# Derivada de la Log-verosimilitud respecto a theta
dl_dtheta = sp.diff(l_theta, theta)
print(f"Derivada de la Log-Verosimilitud l'(theta) igualada a 0:\n0 = {sp.simplify(dl_dtheta)}\n")

# Resolver la ecuación (encontrar la raíz de la derivada)
theta_mle = sp.solve(dl_dtheta, theta)[0]

print(f"El Estimador de Máxima Verosimilitud para theta es: {theta_mle}")
```

## Pregunta 2
Se pretende estimar los valores de producción $Y$ (en miles de toneladas) de cierto material, en función del tiempo transcurrido $X$ (en meses) usando los valores de la tabla:

$$
\begin{array}{|c|c|}
\hline
X & Y \\
\hline
2 & 3 \\
\hline
8 & 51 \\
\hline
14 & 173 \\
\hline
15 & 320 \\
\hline
21 & 576 \\
\hline
\end{array}
$$

Se plantea un modelo de la forma $Y = a + bx + cx^2$. Encontrar los estimadores de mínimos cuadrados para $a$, $b$ y $c$ en este modelo.

```python
# DATOS
X = np.array([2, 8, 14, 15, 21])
Y = np.array([3, 51, 173, 320, 576])

# CONSTRUCCIÓN DE LA MATRIZ DE DISEÑO
X_d = np.column_stack([
    np.ones(len(X)),   # columna para 'a'
    X,                 # columna para 'b'
    X**2               # columna para 'c'
])
print("Matriz de diseño X_d:")
print(X_d)

# Mínimos Cuadrados mediante Ecuaciones Normales
# β̂ = (X_dᵀ X_d)⁻¹ X_dᵀ Y
XtX     = X_d.T @ X_d          # Matriz de información (3x3)
XtY     = X_d.T @ Y            # Vector de proyección (3x1)
beta    = np.linalg.solve(XtX, XtY)   # Más estable que invertir explícitamente
a_hat, b_hat, c_hat = beta

print("\n--- Estimadores de Mínimos Cuadrados ---")
print(f"  â = {a_hat:.4f}")
print(f"  b̂ = {b_hat:.4f}")
print(f"  ĉ = {c_hat:.4f}")

print(f"\nModelo ajustado: Ŷ = {a_hat:.4f} + {b_hat:.4f}·X + {c_hat:.4f}·X²")


# Calulo el SSR
Y_hat = X_d @ beta
residuos = Y - Y_hat
RSS = np.sum(residuos**2)   # Suma de cuadrados de residuos

print(f"\nRSS (suma de cuadrados de residuos) = {RSS:.6f}")
```

## Pregunta 3
Don Francisco tiene 8 clientes a los que les ha vendido mercancías a crédito y, de ellos, 1 están en mora con el pago prometido. Matías, teniendo en cuenta la información disponible, considera que puede modelar el porcentaje $p$ de morosidad según una distribución $\mathcal{B}(1,4)$. Para determinar los parámetros $\alpha$ y $\beta$, decide usar inferencia bayesiana. Con esto, pretende explicarle a Don Francisco, cómo será el comportamiento de pago de sus clientes a crédito. Determinen la distribución a posteriori del parámetro $p$ de porcentaje de morosidad ($\alpha$ y $\beta$). Determinar su media y su varianza.

```python
# 1. Parámetros de la distribución a priori: Beta(alpha_prior, beta_prior)
alpha_prior = 1
beta_prior = 4

print(f"Distribución a Priori: Beta({alpha_prior}, {beta_prior})")

# 2. Datos observados (modelo Binomial)
n = 8  # Número total de clientes
k = 1  # Número de clientes en mora

print(f"Datos observados: n={n}, k={k} (éxitos/mora)")

# 3. Parámetros de la distribución a posteriori: Beta(alpha_post, beta_post)
# Usando la propiedad de conjugación de la familia Beta-Binomial
alpha_post = alpha_prior + k
beta_post = beta_prior + n - k

print(f"\nDistribución a Posteriori: Beta({alpha_post}, {beta_post})")

# 4. Calcular la media y la varianza de la distribución a posteriori
# Fórmulas para una distribución Beta(a, b):
# Media = a / (a + b)
# Varianza = (a * b) / [ (a + b)^2 * (a + b + 1) ]

media_post = alpha_post / (alpha_post + beta_post)

# Numerador y denominador de la varianza para mayor claridad
num_var = alpha_post * beta_post
den_var = ((alpha_post + beta_post)**2) * (alpha_post + beta_post + 1)
varianza_post = num_var / den_var

print("\n--- Resultados Finales ---")
print(f"Parámetros Posteriori: alpha = {alpha_post}, beta = {beta_post}")
print(f"Media esperada del porcentaje de morosidad (p): {media_post:.4f} (aprox {media_post*100:.2f}%)")
print(f"Varianza de p: {varianza_post:.6f}")
```
